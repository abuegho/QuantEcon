### Exercise 1

```python
def factorial(n):
    prod = 1
    for i in range(1, n + 1):
        prod *= i
    return(prod)
```

### Exercise 2

```python
from numpy.random import uniform

def binomial_rv(n, p):
    y = 0
    for i in range(n):  ## Each loop is one Bernoulli trial
        d = uniform()
        if d < p:
            y += 1
    return(y)

binomial_rv(30, .2)
```

### Exercise 3

```python
import numpy as np

x = uniform(-1, 1, 1000000)
y = uniform(-1, 1, 1000000)
ctr = 0
for i in range(1000000):
    rad = (x[i]**2 + y[i]**2)**.5
    if rad < 1:
        ctr += 1
ctr / 250000
```

### Exercise 4

```python
from numpy.random import uniform

heads = 0
for i in range(10):
    if uniform() < .5:
        heads += 1
if heads >= 3:
    print('Pay one dollar')
else:
    print('Pay nothing')
```

### Exercise 5

```python
import numpy as np
import matplotlib.pyplot as plt

x = [np.random.standard_normal()]
α = .9
for i in range(199):
    x.append(α * x[i] + np.random.standard_normal())

plt.plot(x)
plt.show()
```

```python
x = np.random.standard_normal(200)
y = [np.random.standard_normal()]
z = [np.random.standard_normal()]

α = .8
α2 = .98

for i in range(199):
    y.append(α * y[i] + np.random.standard_normal())
    z.append(α2 * z[i] + np.random.standard_normal())
    
plt.plot(x, label = 'α = .0')
plt.plot(y, label = 'α = .8')
plt.plot(z, label = 'α = .98')
plt.legend()
plt.show()
```
